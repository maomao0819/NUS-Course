Note that “seeds_with_headers.csv” was modified from “wheatseeds.csv” downloaded from https://machinelearningmastery.com/implement-backpropagationalgorithm-scratch-python/ Note that there are 8 features arranged in 8 columns i.e. feat1 to feat8. ## import numpy as np

```python
import numpy as np
```

## Use np.mean to compute the mean of each of the 8 features

```python
mean = np.mean(dataset)
print('mean:', mean, sep='\n')
```

## Use np.cov to compute the covariance matrix of the data in the file. hint: set rowvar = False

```python
# covariance = np.cov(dataset, rowvar=False)
covariance = np.cov(dataset - dataset.mean(), rowvar=False)
print('covariance matrix:')
print(covariance)
```

## Use np.linalg.eig to do eigen decomposition of the covariance matrix.

```python
eigenvalue, eiganvector = np.linalg.eig(covariance)
print('eigenvalue:')
print(eigenvalue, '\n')
print('eiganvector:')
print(eiganvector)
```

## Use np.argsort to sort the eigenvalues in descending order

```python
eigenvalue_descend_id = np.argsort(eigenvalue)[::-1]

eigenvalue_descend = eigenvalue[eigenvalue_descend_id]
print('eigenvalue_descend:')
print(eigenvalue_descend, '\n')
eiganvector_descend = eiganvector[eigenvalue_descend_id]
print('eiganvector_descend:')
print(eiganvector_descend, '\n')
```

## Use np.dot to show that the eigenvectors are orthogonal to each other.

```python
order_name = ['1st', '2nd', '3rd'] + [str(i) + 'th' for i in range(4, 8+1)]
for id_1 in range(len(eiganvector_descend)):
    for id_2 in range(id_1 + 1, len(eiganvector_descend)):
        print(f'{order_name[id_1]} eiganvector dot {order_name[id_2]} eiganvector =', np.dot(eiganvector_descend[id_1], eiganvector_descend[id_2]))
    print()
```

Because of the floating-points limitations, the dot values don't equal to 0 but very close to 0. Thus, eigenvectors are orthogonal to each other.

# Q2 Project each data point onto all the eigenvectors and store the results of projection into a matrix coef of size number_of_data_points by 8.

```python
def calculate_coefficient_matrix(observations, eiganvectors, eigan_id=[i for i in range(8)]):
    return np.matmul(np.array(dataset), np.array(eiganvectors)[eigan_id].T)
    # return np.dot(np.array(dataset), np.array(eiganvector_descend)[eigan_id].T)
```

```python
coef = calculate_coefficient_matrix(dataset, eiganvector_descend)
print('coefficient matrix:')
print(pd.DataFrame(data=coef, columns=['eigan_' + str(i) for i in range(1, 8+1)]))
```

## Reconstruct the data points using coef and the eigenvectors.

```python
def reconstruct_data(coefficient_matrix, eiganvectors):
    points = list()
    coef = np.array(coefficient_matrix)
    n_observations = np.shape(coefficient_matrix)[0]
    n_eigans = np.shape(coefficient_matrix)[1]
    for observation in range(n_observations):
        point = np.zeros(8)
        for eigan_id in range(n_eigans):
            point += coef[observation][eigan_id] * eiganvectors[eigan_id]
        points.append(point)
    return points
```

```python
points = reconstruct_data(coef, eiganvector_descend)
print('data points by reconstructing:')
print(pd.DataFrame(data=points, columns=dataset.columns))
```

## Calculate the squared reconstruction errors.

```python
def calculate_squared_reconstruction_errors(data1, data2):
    ## using basic operation
    # reconstruction_errors = np.sum(np.square(points - dataset), axis=1)
    ## using dot
    # reconstruction_errors = [np.dot((points - np.array(dataset))[i], (points - np.array(dataset))[i]) for i in range(np.shape(data1)[0])]
    ## using matrix multiplication
    reconstruction_errors = [np.matmul((points - np.array(dataset))[i].T, (points - np.array(dataset))[i]) for i in range(np.shape(data1)[0])]
    return np.array(reconstruction_errors)

def total_squared_reconstruction_errors(data1, data2):
    reconstruction_errors = calculate_squared_reconstruction_errors(data1, data2)
    return np.sum(reconstruction_errors)
```

```python
points = np.array(points)
reconstruction_errors = calculate_squared_reconstruction_errors(points, dataset)
print('The squared reconstruction error is:')
print(reconstruction_errors, '\n')
total_error = total_squared_reconstruction_errors(points, dataset)
print('The total squared reconstruction errors are:', total_error)
```

## Print the squared reconstruction error for each data point

```python
def print_reconstruction_errors(data1, data2):
    reconstruction_errors = calculate_squared_reconstruction_errors(data1, data2)
    n_observations = np.shape(reconstruction_errors)[0]
    order_name = ['1st', '2nd', '3rd'] + [str(i) + 'th' for i in range(4, n_observations+1)]
    for i in range(len(reconstruction_errors)):
        print(f'The squared reconstruction error of the {order_name[i]} data is:', reconstruction_errors[i])
```

```python
print_reconstruction_errors(points, dataset)
```
